```python
from kraken import blla
from kraken.lib import segmentation, vgsl, xml

MODEL_PATH = "/trunk/danlu/kraken/line_extraction/models/ms_mellon_print_layout.mlmodel"

model = vgsl.TorchVGSLModel.load_model(MODEL_PATH)
```

```python
from glob import glob

all_images = glob("./test_images_240804/*.png")
```

```python
print(all_images)
```

```python
imgurl = "./0533IbnBajja.TacaliqCalaMantiqFarabi_0043.png"
from PIL import Image

image = Image.open(imgurl)
```

```python
MAX_SIZE = 3000
w, h = image.size
scaling_factor = min(MAX_SIZE / w, MAX_SIZE / h)

if scaling_factor < 1:
    new_width = int(w * scaling_factor)
    new_height = int(h * scaling_factor)
    image = image.resize((new_width, new_height), Image.LANCZOS)
print(image.size)
```

```python
res = blla.segment(image, model=model)
```

```python
print(res.keys())
```

```python
from PIL import ImageDraw

def flatten(xy):
    ret = []
    for x, y in xy:
        ret+= [x, y]
    return ret

draw = ImageDraw.Draw(image)

for line in res['lines']:
    # print(line['boundary'])
    draw.polygon(flatten(line['boundary']), outline='red', width=5)
    # break
```

```python
####
from PIL import Image


for img_url in all_images:
    image = Image.open(img_url)
    MAX_SIZE = 3000
    w, h = image.size
    scaling_factor = min(MAX_SIZE / w, MAX_SIZE / h)

    if scaling_factor < 1:
        new_width = int(w * scaling_factor)
        new_height = int(h * scaling_factor)
        image = image.resize((new_width, new_height), Image.LANCZOS)
        
    print(image.size)
    res = blla.segment(image, model=model)
    
    draw = ImageDraw.Draw(image)

    for line in res['lines']:
        # print(line['boundary'])
        draw.polygon(flatten(line['boundary']), outline='red', width=5)
        # break
    image.save(img_url.replace("test_images_240804", "test_iamges_240804_kraken"))
```

```python
import glob
splits = glob.glob(f"/trunk2/shared/datasets/open-iti-2404/HMML/*/*/metadata/*.json")


print(len(splits))
```

```python
import os, random


selected_images = []

for json_url in splits:
    img_folder = os.path.dirname(json_url)
    img_folder = img_folder.replace("metadata", "pageimg")
    images_list = glob.glob(img_folder + "/*")
    try:
        randidx = random.randint(0, len(images_list) - 1)
        selected_images.append(images_list[randidx])
    except:
        continue
    
```
